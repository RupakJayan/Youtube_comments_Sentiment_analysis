```python
import googleapiclient.discovery
from textblob import TextBlob
import matplotlib.pyplot as plt
from collections import Counter
import datetime
import csv

API_KEY = "AIzaSyDuCugD1fxJ8oQSAZBWyHzHhuC_SBByj2U"  # Replace with your YouTube Data API key

def fetch_youtube_comments(video_id, max_results=1000):
    youtube = googleapiclient.discovery.build("youtube", "v3", developerKey=API_KEY)
    comments = []
    nextPageToken = None
    
    while len(comments) < max_results:
        request = youtube.commentThreads().list(
            part="snippet",
            videoId=video_id,
            maxResults=min(100, max_results - len(comments)),
            pageToken=nextPageToken
        )
        response = request.execute()
        comments.extend(response['items'])
        if 'nextPageToken' in response:
            nextPageToken = response['nextPageToken']
        else:
            break
            
    return comments

def analyze_sentiment(comment_text):
    blob = TextBlob(comment_text)
    sentiment_score = blob.sentiment.polarity
    return sentiment_score

def classify_comments(comments):
    good_comments = []
    bad_comments = []
    neutral_comments = []
    
    for item in comments:  
        comment_text = item['snippet']['topLevelComment']['snippet']['textDisplay']
        sentiment_score = analyze_sentiment(comment_text)
        
        if sentiment_score > 0:
            good_comments.append(comment_text)
        elif sentiment_score < 0:
            bad_comments.append(comment_text)
        else:
            neutral_comments.append(comment_text)
            
    return good_comments, bad_comments, neutral_comments

def visualize_sentiment_distribution(good_comments, bad_comments, neutral_comments, title):
    fig, axs = plt.subplots(1, 3, figsize=(18, 6))
    
    labels = ['Good Comments', 'Bad Comments', 'Neutral Comments']
    sizes = [len(good_comments), len(bad_comments), len(neutral_comments)]
    colors = ['lightgreen', 'lightcoral', 'lightskyblue']
    explode = (0.1, 0, 0)
    
    axs[0].pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
    axs[0].axis('equal')
    axs[0].set_title('Distribution of Sentiment Categories - ' + title)
    
    axs[1].hist([analyze_sentiment(comment) for comment in good_comments], bins=20, alpha=0.7, color='lightgreen', label='Good Comments')
    axs[1].hist([analyze_sentiment(comment) for comment in bad_comments], bins=20, alpha=0.7, color='lightcoral', label='Bad Comments')
    axs[1].set_xlabel('Sentiment Score')
    axs[1].set_ylabel('Frequency')
    axs[1].set_title('Comparison of Sentiment Scores - ' + title)
    axs[1].legend(loc='upper right')
    axs[1].grid(True)
    
    axs[2].bar(labels, sizes, color=colors)
    axs[2].set_xlabel('Sentiment Category')
    axs[2].set_ylabel('Number of Comments')
    axs[2].set_title('Number of Comments in Each Sentiment Category - ' + title)
    axs[2].grid(True)
    
    plt.show()

def generate_report(good_comments, bad_comments, neutral_comments, title):
    total_comments = len(good_comments) + len(bad_comments) + len(neutral_comments)
    good_percentage = (len(good_comments) / total_comments) * 100
    bad_percentage = (len(bad_comments) / total_comments) * 100
    neutral_percentage = (len(neutral_comments) / total_comments) * 100
    print(f"Total Comments for {title}: {total_comments}")
    print(f"Good Comments: {len(good_comments)} ({good_percentage:.2f}%)")
    print(f"Bad Comments: {len(bad_comments)} ({bad_percentage:.2f}%)")
    print(f"Neutral Comments: {len(neutral_comments)} ({neutral_percentage:.2f}%)")

    # Count words in good and bad comments
    good_word_count = sum(len(comment.split()) for comment in good_comments)
    bad_word_count = sum(len(comment.split()) for comment in bad_comments)

    print(f"Total Words in Good Comments: {good_word_count}")
    print(f"Total Words in Bad Comments: {bad_word_count}")

    # Calculate sentiment score for each word in good and bad comments
    good_word_sentiment = {}
    for comment in good_comments:
        for word in comment.split():
            if word not in good_word_sentiment:
                good_word_sentiment[word] = analyze_sentiment(word)

    bad_word_sentiment = {}
    for comment in bad_comments:
        for word in comment.split():
            if word not in bad_word_sentiment:
                bad_word_sentiment[word] = analyze_sentiment(word)

    # Display top good words with counts and sentiment scores
    print("Top Good Words with Counts and Sentiment Scores:")
    top_good_words = sorted(good_word_sentiment.items(), key=lambda x: x[1], reverse=True)[:10]
    for word, sentiment_score in top_good_words:
        count = sum(comment.count(word) for comment in good_comments)
        print(f"{word}: Count={count}, Sentiment Score={sentiment_score}")

    # Display top bad words with counts and sentiment scores
    print("Top Bad Words with Counts and Sentiment Scores:")
    top_bad_words = sorted(bad_word_sentiment.items(), key=lambda x: x[1])[:10]
    for word, sentiment_score in top_bad_words:
        count = sum(comment.count(word) for comment in bad_comments)
        print(f"{word}: Count={count}, Sentiment Score={sentiment_score}")

    print()

    return top_good_words, top_bad_words

if __name__ == "__main__":   
    videos = [
        {"video_id": "1cgbDMYsLDw", "title": "PIZZA CHALLENGE"},
        {"video_id": "JjUIJOaPpzA", "title": "Every Indian Restaurant Experience Ever"},
        {"video_id": "p_fXsFfwxTo", "title": "ROAST YOURSELF CHALLENGE - CAELI"},
        {"video_id": "jDZH2Y4fwCc", "title": "3 Ways To Use Aloe Vera Gel To Get Soft, Smooth & Shiny Hair!"},
        {"video_id": "ceKMnyMYIMo", "title": "hi."}
    ]

    for video in videos:
        print(f"Fetching comments for video: {video['title']}")
        comments = fetch_youtube_comments(video['video_id'])
        print(f"Total comments fetched: {len(comments)}")

        good_comments, bad_comments, neutral_comments = classify_comments(comments)

        # Visualize sentiment distribution
        visualize_sentiment_distribution(good_comments, bad_comments, neutral_comments, video['title'])

        # Generate and display report
        generate_report(good_comments, bad_comments, neutral_comments, video['title'])
```
